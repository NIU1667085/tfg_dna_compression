# RESUM MIDA DATASETS

```python
import os

DATA_DIR = "/home/helen/genomic_benchmark/DATA/data"

class EntityInfo:
    name = ""
    total_files = 0
    total_size = 0
    mean_size = 0
    files_needed_to_reach_1GB = 0

summary = []
for directory in os.listdir(DATA_DIR):
    summary.append(EntityInfo())
    summary[-1].name = directory
    total_size = 0
    total_files = 0
    for a in os.listdir(os.path.join(DATA_DIR, directory)):
        total_size += os.path.getsize(os.path.join(DATA_DIR, directory, a))
        total_files += 1

    total_size = total_size / (1024**3)  # Convert bytes to GB
    summary[-1].total_size = total_size 
    summary[-1].total_files = total_files
    summary[-1].mean_size = total_size / total_files if total_files > 0 else 0
    summary[-1].mean_size_GB = summary[-1].mean_size
    summary[-1].files_needed_to_reach_1GB = (1 - total_size) / summary[-1].mean_size 

for entity in summary:
    print(f"{entity.name}:  \t {entity.total_size:.4f} GB  \t {entity.total_files} files  \t mean: {entity.mean_size:.4f} GB/file  \t {entity.files_needed_to_reach_1GB:,.2f} files needed to be added to reach 1GB total size")
```
