```python
import pandas as pd

df1 = pd.read_csv("/home/helen/genomic_benchmark/results/2026_05_26_23_20_45/comparative_results_2026_05_26_23_20_45.csv")
df2 = pd.read_csv("/home/helen/genomic_benchmark/results/2026_05_26_23_19_12/comparative_results_2026_05_26_23_19_12.csv")

cols = [
    "dataset_id",
    "dataset_label",
    "algorithm",
    "original_size_bytes",
    "compressed_size_bytes",
    "compress_seconds",
    "decompress_seconds"
]

df1_small = df1[cols].copy()
df2_small = df2[cols].copy()

df1_small = df1_small.rename(columns={
    "compressed_size_bytes": "compressed_size_bytes_run1",
    "compress_seconds": "compress_seconds_run1",
    "decompress_seconds": "decompress_seconds_run1"
})

df2_small = df2_small.rename(columns={
    "compressed_size_bytes": "compressed_size_bytes_run2",
    "compress_seconds": "compress_seconds_run2",
    "decompress_seconds": "decompress_seconds_run2"
})

key_cols = ["dataset_id", "dataset_label", "algorithm", "original_size_bytes"]

df = df1_small.merge(df2_small, on=key_cols, how="inner")
```
