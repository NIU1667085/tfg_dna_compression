# Diferències entre temps compressió i descompressió

```python
import pandas as pd
results_csv = "/home/helen/genomic_benchmark/scripts/comparative_final_results.csv"
df = pd.read_csv(results_csv)

mean_compress = dict()
mean_decompress = dict()
algos = set(df["algorithm"])

for algo in algos:
    df_algo = df[df["algorithm"] == algo]

    mean_compress[algo] = df_algo["compress_seconds"].mean()
    mean_decompress[algo] = df_algo["decompress_seconds"].mean()

print("Mean times:")
for mean_compress, mean_decompress in zip(mean_compress.items(), mean_decompress.items()):
    algo = mean_compress[0]
    diff = mean_compress[1] - mean_decompress[1]
    print(f"{algo}: {mean_compress[1]:.2f} seconds (compress), {mean_decompress[1]:.2f} seconds (decompress), diff {diff:.2f} seconds ({diff/(mean_compress[1]+mean_decompress[1]):.2%} difference relative to total time)")
```
